EDA:-

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
category_sales
```

```python
plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=category_sales.index,
    y=category_sales.values,
    hue=category_sales.index,   # Gives each bar a different color
    palette="viridis",
    legend=False
)

plt.title("Total Sales by Category", fontsize=16, fontweight="bold")
plt.xlabel("Category", fontsize=12)
plt.ylabel("Sales ($)", fontsize=12)

# Add value labels on top of bars
for i in ax.containers:
    ax.bar_label(i, fmt='%.0f')

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

### Business Insight

The Technology category generated the highest total sales, making it the company's strongest revenue contributor.

Furniture ranked second, while Office Supplies generated the lowest sales among the three categories.

**Recommendation:**
- Continue investing in Technology products.
- Investigate whether Office Supplies needs better marketing or pricing strategies.
- Compare sales with profit before making business decisions, since high sales do not always mean high profitability.

```python
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=category_profit.index,
    y=category_profit.values,
    hue=category_profit.index,
    palette="magma",
    legend=False
)

plt.title("Total Profit by Category", fontsize=16, fontweight="bold")
plt.xlabel("Category")
plt.ylabel("Profit ($)")

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(9,5))

ax = sns.barplot(
    x=region_sales.index,
    y=region_sales.values,
    hue=region_sales.index,
    palette="Set2",
    legend=False
)

plt.title("Total Sales by Region", fontsize=16, fontweight="bold")
plt.xlabel("Region")
plt.ylabel("Sales ($)")

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

### Business Insight

The chart compares total sales across all regions.

- The highest-performing region contributes the largest share of revenue.
- Lower-performing regions may require targeted marketing campaigns or operational improvements.
- Regional analysis helps management allocate budgets and resources effectively.
