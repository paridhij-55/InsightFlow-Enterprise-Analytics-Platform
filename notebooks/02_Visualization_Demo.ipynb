```python
import pandas as pd
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("../data/raw/Superstore.csv", encoding="latin1")

df.head()
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df["Sales"],
    df["Profit"],
    df["Discount"],
    c=df["Profit"],
    cmap="viridis",
    alpha=0.7
)

ax.set_xlabel("Sales")
ax.set_ylabel("Profit")
ax.set_zlabel("Discount")

ax.set_title("3D Scatter Plot: Sales vs Profit vs Discount")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["Sales"],
    df["Profit"],
    c=df["Profit"],
    cmap="viridis",
    alpha=0.7
)

plt.xlabel("Sales")
plt.ylabel("Profit")

plt.title("2D Scatter Plot: Sales vs Profit")

plt.colorbar(label="Profit")

plt.show()
```

```python
sales_category = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(7,5))

sales_category.plot(kind="bar")

plt.title("Sales by Category")

plt.ylabel("Sales")

plt.show()
```
